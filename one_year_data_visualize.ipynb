```python
import matplotlib.pyplot as plt
import rasterio
import numpy as np
import os
```

```python
folder_path = "./output/index/ndvi/2020"
files = os.listdir(folder_path)

target_fl = "_crop.tif"

file_inside = []

for file_name in files:
    if file_name.endswith(target_fl):
            file_inside.append(file_name)
```

```python
file_inside
```

```python
data_mean = []
data_std = []
for f in file_inside:
    with rasterio.open(folder_path + '/' + f) as src:
        data = src.read(1)

    # clear null data
    list = []
    for rows in data:
        for cols in rows:
            if cols != 0:
                list.append(cols)

    data_mean.append(np.mean(list))
    data_std.append(np.std(list))

    src.close()
```

```python
print(data_mean)
print(data_std)
```

```python
# Create a list of time points for the x-axis (you can customize these if needed)
time_points = range(1, len(data_mean) + 1)

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(time_points, data_mean, marker='o', linestyle='-')

# Set labels and title
plt.xlabel('Month')
plt.ylabel('NDVI')
plt.title('Year 2020 - NDVI Value')

# Display the plot
plt.xticks(time_points)
plt.tight_layout()
plt.show()
```
